```python
import requests
page = requests.get("https://www.math.uni.lodz.pl/lista-pracownikow/")
page
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('td')
```

```python
titles = soup.find_all(class_="tytul-field")
titles
```

```python
firstNames = soup.find_all(class_="first_name-field")
firstNames
```

```python
lastNames= soup.find_all(class_="last_name-field")
```

```python
links = soup.findAll(class_="last_name-field")
```

```python
emails = soup.find_all(class_="email-field")
emails
```

```python
websiteLists = []
roomList = []
for link in links: 
    page = requests.get(link.findChild("a")["href"])
    soup = BeautifulSoup(page.content, 'html.parser')
    website = soup.find(class_="pdb-website link")
    room = soup.find(class_="pdb-pokoj text-line")
    if website != None: 
        websiteLists.append(website.findChild("dd").findChild("a")["href"])
    else: 
        websiteLists.append("None")
    if room != None: 
        roomList.append(room.findChild("dd").getText())
    else: 
        roomList.append("None")
```

```python
titlesList = []
for title in titles: 
    titlesList.append(title.get_text().strip())

firstNameList = []

for name in firstNames: 
    firstNameList.append(name.get_text().strip())

lastNameList = []

for lastName in lastNames: 
    lastNameList.append(lastName.get_text().strip())
    
emailList = []

for email in emails: 
    emailList.append(email.get_text().strip())
```

```python
from pandas import DataFrame

professors = DataFrame(titlesList, columns=['title'])
professors['professor'] = firstNameList
professors['lastName'] = lastNameList
professors['email'] = emailList
professors['room'] = roomList
professors["website"] = websiteLists
```

```python
print(professors)
```

```python
professors.to_csv("professors.csv", index=False,sep=" ")
```
